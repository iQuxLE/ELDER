```python
from collections import Counter

def get_duplicate_keys(dictionary):
    counts = Counter(dictionary.keys())
    return [key for key, count in counts.items() if count > 1]

def more_than_one_key(keys):
    count = {}
    for k in keys:
        count[k] = count.get(k, 0) + 1
    return [k for k, v in count.items() if v > 1]

# Example usage
duplicates = get_duplicate_keys(hp_embeddings_dict)
keys = hp_embeddings_dict.keys()
duplicates_2 = more_than_one_key(keys=keys)
```
